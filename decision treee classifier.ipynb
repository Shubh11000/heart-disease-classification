#### SOME CATEGORIES ONLY HAVE TWO UNIQUE VALUES, SO WE'LL CHANGE THOSE CATEGORIES TO HAVE EITHER A 1 OR A 0

```python
df['HeartDiseaseYes'] = df['HeartDisease'].map({'Yes':1,'No':0})
df['SmokingYes'] = df['Smoking'].map({'Yes':1,'No':0})
df['AlcoholDrinkingYes'] = df['AlcoholDrinking'].map({'Yes':1,'No':0})
df['StrokeYes'] = df['Stroke'].map({'Yes':1,'No':0})
df['DiffWalkingYes'] = df['DiffWalking'].map({'Yes':1,'No':0})
df['SexMale'] = df['Sex'].map({'Male':1,'Female':0})
df['PhysicalActivityYes'] = df['PhysicalActivity'].map({'Yes':1,'No':0})
df['AsthmaYes'] = df['Asthma'].map({'Yes':1,'No':0})
df['KidneyDiseaseYes'] = df['KidneyDisease'].map({'Yes':1,'No':0})
df['SkinCancerYes'] = df['SkinCancer'].map({'Yes':1,'No':0})
```

```python
df=df.drop(['HeartDisease','Smoking','AlcoholDrinking','Stroke','DiffWalking','Sex','PhysicalActivity','Asthma','KidneyDisease','SkinCancer'], axis=1)
```

```python
df.head()
```

#### SOME CATEGORIES HAVE MORE THAN 2 NON-NUMERICAL VALUES, WE'LL SPLIT EACH VALUE INTO ITS OWN CATEGORY

```python
dummies = pd.get_dummies(df[['AgeCategory', 'Race', 'Diabetic', 'GenHealth']],drop_first=True)
df = df.drop(['AgeCategory', 'Race', 'Diabetic', 'GenHealth'],axis=1)
df = pd.concat([df,dummies],axis=1)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('HeartDiseaseYes',axis=1)
y = df['HeartDiseaseYes']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```
